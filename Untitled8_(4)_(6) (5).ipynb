```python
class CNNEncoder(nn.Module):
  def __init__(self, embedding_dim = 20):
    super(CNNEncoder, self).__init__()
    self.conv1 = nn.Sequential(
        nn.Conv2d(1,6,2,padding = 1),
        nn.BatchNorm2d(6),
        nn.ReLU(),
        nn.MaxPool2d((2,2), padding=1)
    )
    self.conv2 = nn.Sequential(
        nn.Conv2d(6,8,3, padding = 0, stride=2),
        nn.BatchNorm2d(8),
        nn.ReLU(),
        nn.MaxPool2d((2,2), padding=1)
    )
    self.linear = nn.Sequential(
        nn.Flatten(start_dim=1),
        nn.Linear(2 * 2 *8, embedding_dim)
    )
  def forward(self, x):
    print('BEFORE CONV1')
    print(x.shape)
    x = self.conv1(x)
    print('AFTER CONV1')
    print(x.shape)
    x = self.conv2(x)
    print('FINAL ENCODER SHAPE')
    print(x.shape)
    x = self.linear(x)
    print('FINAL ENCODER')
    print(x.shape)
    return x

class CNNDecoder(nn.Module):
  def __init__(self, embedding_dim):
    super(CNNDecoder, self).__init__()
    self.deconv1 = nn.Sequential(
        nn.Linear(embedding_dim, 2*2*8),
        nn.Unflatten(dim=1, unflattened_size=(8, 2, 2)),
        nn.ConvTranspose2d(8, 6, 2, stride=2, dilation=2),
        nn.BatchNorm2d(6),
        nn.ReLU(True)
    )
    self.deconv2 = nn.Sequential(
        nn.ConvTranspose2d(6, 1, 2, output_padding=1, dilation=2),
        #nn.BatchNorm2d(6),
        #nn.ReLU(True)
    )
    self.deconv3 = nn.Sequential(
        nn.ConvTranspose2d(6, 1, 2, padding = 1)
    )
  def forward(self, x):
    print('DECONV1 BEFORE SHAPE')
    print(x.shape)
    x = self.deconv1(x)
    print('DECONV2 BEFORE SHAPE')
    print(x.shape)
    x = self.deconv2(x)
    print('DECONV3 BEFORE SHAPE')
    print(x.shape)
    #x = self.deconv3(x)
    # print('DECONV3 AFTER SHAPE')
    # print(x.shape)
    return x
```

```python
class CNNAutoencoder(nn.Module):
  def __init__(self, embedding_dim=20):
    super(CNNAutoencoder, self).__init__()
    self.encoder = CNNEncoder(embedding_dim).to(device)
    self.decoder = CNNDecoder(embedding_dim).to(device)
  def forward(self, x):
    x = self.encoder(x)
    x = self.decoder(x)
    return x
```

```python
# embedding_dim=20
# hidden_size_l1=32
# hidden_size_l2=16
# hidden_size_l3=8
class LSTMautoencoder(nn.Module):
  def __init__(self, nbre_features, hidden_size_l1, hidden_size_l2, hidden_size_l3):#, seq_len):
    super(LSTMautoencoder, self).__init__()
    #self.seq_len = seq_len
    self.nbre_features = nbre_features
    self.hidden_size_l1 = hidden_size_l1
    self.hidden_size_l2 = hidden_size_l2
    self.hidden_size_l3 = hidden_size_l3

    self.LSTM_EL1 = nn.LSTM(
        input_size = nbre_features,
        hidden_size = hidden_size_l1,
        num_layers = 1,
        batch_first = True
    )
    self.LSTM_EL2 = nn.LSTM(
        input_size = hidden_size_l1,
        hidden_size = hidden_size_l2,
        num_layers = 1,
        batch_first = True
    )
    self.LSTM_EL3 = nn.LSTM(
        input_size = hidden_size_l2,
        hidden_size = hidden_size_l3,
        num_layers = 1,
        batch_first = True
    )

    self.LSTM_DL1 = nn.LSTM(
        input_size = hidden_size_l3,
        hidden_size = hidden_size_l2,
        num_layers = 1,
        batch_first = True
    )
    self.LSTM_DL2 = nn.LSTM(
        input_size = hidden_size_l2,
        hidden_size = hidden_size_l1,
        num_layers = 1,
        batch_first = True
    )
    self.LSTM_DL3 = nn.LSTM(
        input_size = hidden_size_l1,
        hidden_size = nbre_features,
        num_layers = 1,
        batch_first = True
    )
    #self.LINEAR = nn.Linear(hidden_size, nbre_features)
  
  def forward(self, x):
    seq_len = x.shape[1]
    x = x.reshape((1, seq_len, self.nbre_features))
    x, (_,_) = self.LSTM_EL1(x)
    x, (_,_) = self.LSTM_EL2(x)
    x, (h,_) = self.LSTM_EL3(x)
    x        = h.repeat(1, seq_len, 1)
    x, (h,_) = self.LSTM_DL1(x)
    x, (h,_) = self.LSTM_DL2(x)
    x, (h,_) = self.LSTM_DL3(x)
    return x.reshape((seq_len, self.nbre_features))

```

```python
def predict(CNNmodel, LSTMmodel, data):
  losses = []
  criterion = nn.L1Loss(reduction='sum').to(device)
  DS = DataLoader(dataset=custom_Dataset(data),
                      batch_size=1,
                      collate_fn=collate_function,
                      pin_memory = True)
  with torch.no_grad():
    CNNmodel.eval()
    LSTMmodel.eval()
    for step, batch in enumerate(DS):
        subValid = DataLoader(dataset=CNN_custom_intermediary_dataset(batch.image),
                        batch_size = 4)
        imageTensors = []
        for index, image in enumerate(subValid):
          sequence = image.to(device, dtype=torch.float)
          pred = CNNmodel(sequence)
          imageTensors.append(pred)
        final = torch.cat(imageTensors, dim=0)
        tens = torch.unsqueeze(torch.cat((batch.breath.to(device, dtype=torch.float), final), dim=-1), 0)
        sequence = tens.to(device, dtype=torch.float)
        pred = LSTMmodel(sequence)
        loss_step = criterion(torch.unsqueeze(pred, 0), sequence)
        losses.append(loss_step.item())
  return losses
```
